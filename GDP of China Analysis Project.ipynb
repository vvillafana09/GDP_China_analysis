```python
# Creating a plot to visualize the change in GDP in China over the past decade
china_code = codes[codes['Name'] == 'China']['Code']
china_gdp_plot = gdp_data[gdp_data['Country Code'].isin(china_code)]
china_gdp_plot = china_gdp_plot[(china_gdp_plot['Year'] >= 2006) & (china_gdp_plot['Year'] <= 2016) ]
```

```python
china_gdp_plot = china_gdp_plot.set_index('Year')
china_gdp_plot
```

```python
china_gdp_plot.plot(title='China GDP', xlabel='Year', ylabel='GDP Value')
```

```python
china_code = codes[codes['Name'] == 'China']['Code']
china_gdp_growth_year = gdp_data[gdp_data['Country Code'].isin(china_code)]
```

```python
china_gdp_growth_year['Yearly_Growth'] = china_gdp_growth_year['Value'].pct_change() * 100
china_gdp_growth_year
```

```python
china_gdp_growth_year = china_gdp_growth_year.set_index('Year')
china_gdp_growth_year
```

```python
china_gdp_growth_year['Yearly_Growth'].plot(title='China Yearly GDP Growth', xlabel='Year', ylabel='GDP Growth')
```

```python
china_code = codes[codes['Name'] == 'China']['Code']
china_gdp_growth_decade = gdp_data[gdp_data['Country Code'].isin(china_code)]
```

```python
china_gdp_growth_decade['Decade_Growth'] = china_gdp_growth_decade['Value'].pct_change(10) * 100
china_gdp_growth_decade
```

```python
decades = [1960, 1970, 1980, 1990, 2000, 2010, 2016]
china_gdp_growth_decade = china_gdp_growth_decade[china_gdp_growth_decade['Year'].isin(decades)]
china_gdp_growth_decade
```

```python
china_gdp_growth_decade = china_gdp_growth_decade.set_index('Year')
china_gdp_growth_decade
```

```python
china_gdp_growth_decade['Decade_Growth'].plot(title='China Decade GDP Growth', xlabel='Year', ylabel='GDP Growth')
```
